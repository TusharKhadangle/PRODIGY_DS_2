```python
import pandas as pd

df=pd.read_csv('breast_cancer_survival.csv')
df
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
# it is seen that there are Nan values in the  Date_of_Last_Visit ,Patient_Status column so we will clean it.
```

```python
# since it is not that important column for our analysis ,Date of last visit we will be  just excluding the rows

df = df.dropna(subset=['Date_of_Last_Visit'])
df
```

```python
# since Patient_Status is an important variable for our analysis i am not removing any rows
# i am using mode imputation method. for data cleaning
# I am replacing missing values with the mode (the most frequently occuring varaiable) in the column

most_frequent_status = df['Patient_Status'].mode()[0]
df['Patient_Status'].fillna(most_frequent_status, inplace=True)
```

```python
# Now the data is cleaned there are no NAN values in the data
print(df.isnull().sum())
```

```python
# we are t aking the Min age to understand and analyze what can be the minumum age from the dataset to have breast cancer
age_column = df['Age']

min_age=age_column.min()

print("Minimum age",min_age)
```

```python
age_column = df['Age']

max_age=age_column.max()

print("Maximum age",max_age)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# this shows us the death ratio how many are alive how many are dead
sns.countplot(x='Patient_Status', data=df)
plt.title('People survived having breast cancer')
plt.show()
```

```python
# this chart is showing us the patientstatus dead or alive according to tumourstage

sns.countplot(x='Tumour_Stage', hue='Patient_Status', data=df)
plt.title('Survival according to Tumour stage ')
plt.show()
```

```python
sns.countplot(x='Age', hue='Patient_Status', data=df)
plt.title('Survival according to Age ')
plt.show()
```

```python
sns.countplot(x='Gender',hue="Patient_Status", data=df)
plt.title('People survived having breast cancer')
plt.show()
```

```python
sns.countplot(x='Gender', data=df)
plt.title('Genders having breast cancer')
plt.show()
```

```python
sns.boxplot(x='Tumour_Stage', y='Age', data=df)
plt.title('Tumour stage and age')
plt.show()
```

```python

# this will give us the view on the flow of age of people having breast cancer
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=50, color='lightblue', label="Age", kde=True)
plt.legend()
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Age', y='Tumour_Stage', data=df, inner='quartile')
plt.title('Violin Plot of Age and tumour stage')
plt.xlabel('Target ( Age)')
plt.ylabel('Tumour stage')
plt.show()
```

```python
sns.kdeplot(x='Age',hue='Patient_Status' ,data=df)
plt.title('People survived having breast cancer')
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
